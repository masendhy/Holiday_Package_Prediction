```python
# import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')
```

```python
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from matplotlib import rcParams

rcParams['figure.figsize'] = 10, 8
sns.set_style('whitegrid')
rcParams['font.family'] = 'sans-serif'
rcParams['lines.linewidth'] = 2.5
rcParams['xtick.labelsize'] = 15
rcParams['ytick.labelsize'] = 15
```

```python
# read data
df = pd.read_csv('dataset/Travel.csv')
```

### **Handling Missing Values**

```python
# remove missing value from column TypeofContact, NumberOfFollowups, PreferredPropertyStar, and NumberOfTrips

df.dropna(subset=['TypeofContact','NumberOfFollowups','PreferredPropertyStar','NumberOfTrips'], inplace=True)

# fill missing value with mean from column Age, DurationOfPitch, MonthlyIncome

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['DurationOfPitch'].fillna(df['DurationOfPitch'].mean(), inplace=True)
df['MonthlyIncome'].fillna(df['MonthlyIncome'].mean(), inplace=True)

# lost data 235
```

```python
# drop column CustomerID and Number of children visiting

df.drop(['CustomerID','NumberOfChildrenVisiting'], axis=1, inplace=True)
```

```python
# lets transform float type data to int

df = df.astype({'Age': 'int64', 'DurationOfPitch': 'int64', 'MonthlyIncome': 'int64','NumberOfFollowups': 'int64', 'NumberOfTrips': 'int64', 'PreferredPropertyStar': 'int64'})
```

```python
df.info()
```

### **Handling Unmatched Data Value**

```python
# replace Fe Male into Female in column Gender 

df['Gender'] = df['Gender'].replace(['Fe Male'],'Female')
```

```python
df['Gender'].value_counts()
```

```python
# replace Unmarried and Divorce into Single in column MaritalStatus

df['MaritalStatus'] = df['MaritalStatus'].replace(['Unmarried','Divorced'],'Single')
```

```python
df['MaritalStatus'].value_counts()
```

```python
# check duplicate data

df.duplicated().sum()
```

```python
# check data value

categorical_df = pd.DataFrame()
for col in df.select_dtypes(include=['object','int64']):
    categorical_df[col] = df[col]
for col in categorical_df:
    print(f'{col :-<50} {df[col].unique()}')
```

### **Handing Outliers**

```python
# before we do outlier detection, we need to check the distribution of data

fig, ax = plt.subplots(2, 3, figsize=(20, 10))
sns.distplot(df['Age'], ax=ax[0][0])
sns.distplot(df['DurationOfPitch'], ax=ax[0][1])
sns.distplot(df['NumberOfFollowups'], ax=ax[0][2])
sns.distplot(df['MonthlyIncome'], ax=ax[1][1])
sns.distplot(df['NumberOfTrips'], ax=ax[1][2])
sns.distplot(df['PreferredPropertyStar'], ax=ax[1][0])
plt.show()
```

```python
# count column before outlier detection

print(f'Number of rows before filtering outliers:{len(df)}')

# outlier detection

i = ['Age', 'DurationOfPitch', 'NumberOfFollowups', 'MonthlyIncome', 'DurationOfPitch','NumberOfFollowups'  ]

#outlier detection using Quantile

Q1 = df[i].quantile(0.25)
Q3 = df[i].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - (1.5*IQR)
upper_limit = Q3 + (1.5*IQR)

df_fil = df[~((df[i] < lower_limit) | (df[i] > upper_limit)).any(axis=1)]

# outlier detection using Z-Score

# z = np.abs(stats.zscore(df[i]))
# df_fil = df[(z < 3).all(axis=1)]


print(f'Number of rows after filtering outliers using Quantile:{len(df_fil)}')
```

```python
# show data distribution after outlier detection

fig, ax = plt.subplots(2, 3, figsize=(20, 10))
sns.distplot(df_fil['Age'], ax=ax[0][0])
sns.distplot(df_fil['DurationOfPitch'], ax=ax[0][1])
sns.distplot(df_fil['NumberOfFollowups'], ax=ax[0][2])
sns.distplot(df_fil['MonthlyIncome'], ax=ax[1][1])
sns.distplot(df_fil['NumberOfTrips'], ax=ax[1][2])
sns.distplot(df_fil['PreferredPropertyStar'], ax=ax[1][0])
plt.show()
```

### **Future Transformation**

```python
# rescailing data

i = ['Age' ]

df_fil['scaler'] = MinMaxScaler().fit_transform(df_fil[i].values.reshape(len(df_fil), 1))

df_fil['stand'] = StandardScaler().fit_transform(df_fil[i].values.reshape(len(df_fil), 1))

df_fil.describe()
```

### **Feature Encoding**

```python
# transform categorical data into numerical data

df_fil['Gender_num'] = df_fil['Gender'].apply(lambda x : 1 if x == 'Male' else 0)
df_fil['MaritalStatus_num'] = df_fil['MaritalStatus'].apply(lambda x : 1 if x == 'Married' else 0)
df_fil['TypeofContact_num'] = df_fil['TypeofContact'].apply(lambda x : 1 if x == 'Self Enquiry' else 0)

df_fil.sample(3)
```

```python
# droping unnecessary column

df_new = df_fil.drop(columns=['CityTier','Occupation','Gender','NumberOfPersonVisiting','ProductPitched','PitchSatisfactionScore','OwnCar','Designation','PreferredPropertyStar','NumberOfTrips']).copy()
```

```python
# show correlation between data

plt.figure(figsize=(20,20))
sns.heatmap(df_new.corr(), annot=True, cmap='Oranges',fmt='.2f')
plt.show()
```
